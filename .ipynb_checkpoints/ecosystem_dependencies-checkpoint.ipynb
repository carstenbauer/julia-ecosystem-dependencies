# Parse dependencies

```julia
using DependenciesParser

const ALLPKGS = DependenciesParser.data
const stdlibs = ["Base64", "CRC32c", "Dates", "DelimitedFiles",
 "Distributed", "FileWatching", "Future", "InteractiveUtils",
  "LibGit2", "Libdl", "LinearAlgebra", "Logging", "Markdown",
   "Mmap", "Printf", "Profile", "REPL", "Random", "SHA", "Serialization",
    "SharedArrays", "Sockets", "SparseArrays", "Statistics", "SuiteSparse",
     "Test", "UUIDs", "Unicode", "Pkg"]
getdeps(pkg::AbstractString) = installable(pkg, direct=false)[2]

function getalldeps(pkgs=ALLPKGS)
    deps = Dict{String,Vector{String}}()
    @time for (idx, pkg) ∈ enumerate(pkgs)
        println(idx)
        deps[pkg] = getdeps(pkg)
    end
    deps
end
```

# Load from Disk

```julia
using BSON

deps = BSON.load("alldeps.bson")
# drop stdlibs
deps = Dict(k => setdiff(v, stdlibs) for (k,v) in deps if !(k in stdlibs))
# Remove self-dependencies
for (k,v) in deps
    k in v && filter!(!isequal(k), v)
end
```

```julia
function create_graph(pkgs, deps)
#     ps = sort(pkgs)
    ps = pkgs
    pkgs_deps = filter(p -> first(p) in ps, deps)
    nv = length(ps)

    # Build adjacency_matrix, A_ij, where A_ij = 1 indicates an edge from j to i
    am = zeros(Int64, nv, nv)
    for (k,v) in pkgs_deps
        j = findfirst(isequal(k), ps)
        for dep in v
            i = findfirst(isequal(dep), ps)
            if !isnothing(i)
                am[i,j] = 1
            end
        end
    end
    g = DiGraph(am)
end
```

```julia
pkgs = ["StatsBase", "GLM", "StatsModels", "DataFrames", "StatsFuns", "Distributions", "Clustering", "MultivariateStats", "MLBase", "KernelDensity"]
# pkgs = ["Flux", "GLM", "StatsModels", "DataFrames"]
# pkgs = collect(keys(deps)) # all pkgs

g = create_graph(pkgs, deps)
gp = gplot(g, layout=circular_layout, nodelabel=pkgs)
```

# Reverse Dependencies

```julia
function who_depends_on(pkg::AbstractString, deps)
    r = String[]
    for (k,v) in deps
        pkg in v && push!(r, k)
    end
    !isempty(r) ? r : nothing
end
```

```julia
who_depends_on("SortingAlgorithms", deps)
```

```julia
who_depends_on("StatsBase", deps)
```

```julia
who_depends_on("SortingAlgorithms", deps)
```

```julia
revdeps = Dict{String, Vector{String}}()
for pkg in keys(deps)
    l = who_depends_on(pkg, deps)
    if !isnothing(l)
        revdeps[pkg] = l
    end
end
```

```julia
using DataStructures
rdep = OrderedDict{String, Int}()

vs = values(revdeps) .|> length
idxs = sortperm(vs, rev=true)
sort!(vs, rev=true)
ks = collect(keys(revdeps))[idxs]

for i in 1:length(ks)
   rdep[ks[i]] = vs[i]
end
```

```julia
rdep
```

```julia
function add_annotations(p, rdep, pkgs)
    for pkg in pkgs
        idx = findfirst(isequal(pkg), ks)
        n = rdep[pkg]
        annotate!(p, idx, n+20, Plots.text(pkg*".jl", 10, :left))
    end
    nothing
end
```

```julia
collect(keys(rdep))[100:120]
```

```julia
p = Plots.plot(values(rdep) |> collect, color=:green, ylab="# of reverse dependencies", xlab="package", legend=false)
annotate!(p, 400, 950, Plots.text("Total # of packages with rev deps: $(length(rdep))", 10, :left))
annotate!(p, 400, 880, Plots.text("Avg. # of rev deps: $(round(mean(values(rdep)), digits=1))", 10, :left))

add_annotations(p, rdep, ["Compat", "BinDeps", "DataStructures", "StaticArrays", "StatsBase", "Parameters", "FileIO", "PyCall", "Colors"])
savefig("revdeps.pdf")
savefig("revdeps.png")
p
```

```julia
function my_shell_layout(g, nlist::Union{Nothing, Vector{Vector{Int}}} = nothing)
    if GraphPlot.nv(g) == 1
        return [0.0], [0.0]
    end
    if nlist == nothing
        nlist = [collect(1:nv(g))]
    end
    radius = 0.0
    if length(nlist[1]) > 1
        radius = 1.0
    end
    locs_x = Float64[]
    locs_y = Float64[]
    for nodes in nlist
        # Discard the extra angle since it matches 0 radians.
        θ = range(0, stop=2pi, length=length(nodes)+1)[1:end-1]
        append!(locs_x, radius*cos.(θ))
        append!(locs_y, radius*sin.(θ))
        radius *= 4.0
    end
    return locs_x, locs_y
end
```

```julia
use = ["Flux", "DifferentialEquations", "JuMP", "Optim", "DataFrames", "Makie", "DynamicalSystems", "Images", "GLM", "Luxor"]
core = ["DataStructures", "StaticArrays", "OrderedCollections", "Colors", "StatsBase"]
pkgs = vcat(core, use)

g = create_graph(pkgs, revdeps)
```

```julia
gp = gplot(g, layout=spring_layout, nodelabel=pkgs)
```

```julia
nlist = Vector{Vector{Int}}(undef, 2) # two shells
nlist[1] = 1:length(core)
nlist[2] = 1:length(use)
locs_x, locs_y = my_shell_layout(g, nlist)

gp = gplot(g, locs_x, locs_y, nodelabel=pkgs, NODELABELSIZE=2, NODESIZE=0.1, EDGELINEWIDTH=1, nodestrokelw=1, nodestrokec=colorant"grey")
```

```julia
draw(PDF("revdeps_graph.pdf", 25cm, 16cm), gp)
```
